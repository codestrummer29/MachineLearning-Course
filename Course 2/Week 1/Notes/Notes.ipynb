# Minimizing/maximizing the function 

concave/convex function
• concave - line always below the curve
• convex - line always above the curve
• neither - line crosses the function/ stays both above and below the curve

Use hill descent method and find derivate at every point to know max or min value of the function

# Choosing stepsize to reach optimum

decrease with number of iterations

for convex - optimum occurs when derivative is 0

in practice stop when |derivative| < E
